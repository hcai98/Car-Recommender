```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
input_path = '../../data/clean_cars.csv'
output_path = '../../data/feature.csv'
feature_cols = ['Engin_size', 'Gearbox', 'Seat_num', 'Door_num']
is_get_dummies = True
is_standardize = True
```

```python
data = pd.read_csv(input_path)

feature = data[feature_cols]

if is_get_dummies:
    feature = pd.get_dummies(data[feature_cols]).astype(float)

if is_standardize:
    scaler_std = StandardScaler()
    feature = pd.DataFrame(scaler_std.fit_transform(feature), columns=feature.columns)

feature.to_csv(output_path, index=False)
```
