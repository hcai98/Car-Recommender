## Preprocess Data

```python
from typing import Union, Callable
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_path = '../../data/raw/table/Ad_table.csv'
output_path = '../../data/clean_cars.csv'
key_path = '../../data/keys.csv'

# variables to be transformed
vars_strip_numeric = ['Engin_size']
vars_drop_non_numeric_rows = ['Price']

# for aggregations
key_cols = ['Maker', ' Genmodel', 'Year', 'Bodytype']
agg_cols_transforms = {' Genmodel_ID': 'first',
                       'Engin_size':'first', 
                       'Gearbox': 'mode',
                       'Fuel_type': 'first',
                       'Price': 'mean', 
                       'Seat_num': 'mode',
                       'Door_num': 'mode'}

# rename
rename_map = {'Adv_year': 'Year'}

new_index = 'Car_ID'
```

```python
# map method strings to functions
def method_to_func(method_string: str) -> Union[Callable, str]:
    if method_string == 'mean':
        return 'mean'
    elif method_string == 'first':
        return 'first'
    elif method_string == 'mode':
        # return pd.Series.mode
        return lambda x: pd.Series.mode(x)[0]
```

```python
# read data from input path
df = pd.read_csv(input_path)

# rename original columns
df = df.rename(columns=rename_map)

# strip the nuemrical engine size from string
for var_strip_numeric in vars_strip_numeric:
    df[var_strip_numeric] = (df[var_strip_numeric].
                             str.replace('[^\d.]', '')
                             .astype(float))

# drop rows that doesn't have numeric price
for var_drop_non_numeric_rows in vars_drop_non_numeric_rows:
    df = df[df[var_drop_non_numeric_rows].astype(str).str.isnumeric()]
    df[var_drop_non_numeric_rows] = df[var_drop_non_numeric_rows].astype(int)

# drop nas
df = df.dropna()

# map transform methods to transform functions
agg_cols_transforms_funcs = {k: method_to_func(v) for k, v in agg_cols_transforms.items()}

# aggregate raw data by keys
df_output = (
    df
    .groupby(key_cols)
    .agg(agg_cols_transforms_funcs)
    .reset_index()
    .reset_index(drop=False)
    .rename(columns={'index':new_index})
)


# archive key_columns
pd.Series(key_cols).to_csv(key_path, index=False)

display(df_output)

df_output.to_csv(output_path, index=False)
```
