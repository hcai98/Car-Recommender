```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
prediction_path = '../../data/predictions.csv'
key_path = '../../data/keys.csv'
feature_path = '../../data/feature.csv'
user_input = ['BMW', '3 Series', 2018, 'Saloon']

class_col = 'class'
display_cols = ['']
```

```python
# get the prediciton tables with cluster assignment
prediction = pd.read_csv(prediction_path)

# retrieve the key of the prediction column
key_cols = pd.read_csv(key_path).values.reshape(-1).tolist()

# retrieve the feature data
feature_cols =  pd.read_csv(feature_path).columns.to_numpy().tolist()


display(prediction)
display(key_cols)
feature_cols
```

```python
df_input = prediction.copy()

# get information of the input car
for i in range(len(key_cols)):
    df_input = df_input[df_input[key_cols[i]] == user_input[i]]
if len(df_input) == 0:
    raise ValueError("No such car")

# get cluster of input
input_class = df_input[class_col].iloc[0]

df_input
```

```python
# get top 10 similar cars
df_similar_cars = prediction[prediction[class_col] == input_class].sample(10, random_state=1234)
df_similar_cars
```
